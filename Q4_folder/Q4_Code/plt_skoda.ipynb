```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
from scipy import stats
```

```python
# Read csv file
plt_skoda_df = Path("../Input/skoda_data.csv")
df = pd.read_csv(plt_skoda_df)
df.head()
```

```python
#Setting up a, b for correlation coefficient test
a, b = np.polyfit(df['power_ps'], df['price_in_euro'], 1)
```

```python
#Plotting Graph (Scatter, x&y labels, title, best fit line)
plt.scatter(df['power_ps'], df['price_in_euro'], marker="o", facecolors="yellow", edgecolors="black",
            alpha=0.5)
plt.xlabel("power(ps)")
plt.ylabel("price(euro)")
plt.title("Skoda Power and Price Relations")
plt.plot(df['power_ps'], a*df['power_ps']+b)
plt.savefig('../Output/Skoda_Graph.png')
```

```python
# Pearson's Correlation Test
df['price_in_euro'].corr(df['power_ps'])
```
