```python
# import the package we need
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the clean data.
df_all = pd.read_csv('base_data.csv')
df_Five = pd.read_csv('base_5_data.csv')
df_all.head()
```

```python
# check the information of each coloumn
df_all.info()
```

```python
# Convert the'year' coloumn into string.
df_all.loc[:, 'year'] = df_all['year'].astype(str)
df_Five.loc[:, 'year'] = df_Five['year'].astype(str)
```

```python
# plot the scatter plot to check whether have relationship between year and price
plt.scatter(df_all['year'], df_all['price_in_euro'],marker="o")
plt.title('Car Prices Over the Years')
plt.xlabel('Year')
plt.xticks(rotation="vertical")
plt.ylabel('Car Price (Euro)')
plt.show()
```

```python
# calculate the average price of the used car by year
yearly_avg_prices = df_all.groupby('year')['price_in_euro'].mean()
yearly_avg_prices
```

```python
# plot the ling graph of average price of car Vs years(from 1995 to 2023)
plt.figure(figsize=(12, 6))
yearly_avg_prices.plot(marker='o')
plt.title('Average Car Prices Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price (Euro)')
plt.savefig('output/Average_Prices_Years.png')
plt.show()
```
