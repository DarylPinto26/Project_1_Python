```python
import pandas as pd
from pathlib import Path
```

```python
data = Path("data.csv")
```

```python
og_data = pd.read_csv(data)
og_data.head()
```

```python
# Columns we're interested in:
og_data = og_data[['brand','model','color','registration_date','year','price_in_euro','power_ps',
                   'transmission_type','fuel_type','fuel_consumption_l_100km','mileage_in_km']]
og_data
```

```python
og_data.dtypes
```

```python
og_data.count()
```

```python
og_data1 = og_data.dropna()
og_data1.count()
```

```python
og_data1.head(20)
```

```python
og_data1['fuel_consumption_l_100km']
```

```python
og_data1['fuel_consumption_l_100km'] = og_data1['fuel_consumption_l_100km'].str.replace(',', '').str.extract(r'(\d+\.\d+|\d+)').astype(float)
```

```python
og_data1['fuel_consumption_l_100km']
```

```python
# Dropping numeric values from the fuel_type 
og_data1 = og_data1[~og_data1['fuel_type'].str.contains(r'\d')]
```

```python
# Dropping other unwanted values from the fuel_type 
values_to_drop = ['Automatic', 'Manual', 'Other', 'Unknown']

og_data1 = og_data1[~og_data1['fuel_type'].isin(values_to_drop)]
```

```python
# Dropping rows with unknown values from the transmission_type 
values_to_drop1 = ['Unknown']

og_data1 = og_data1[~og_data1['transmission_type'].isin(values_to_drop1)]
```

```python
og_data1.to_csv('clean_data2.csv', index=False)
```

```python
# Dropping rows with un-realistic fuel consumption figures from the column fuel_consumption_l_100km

og_data1 = og_data1[og_data1['fuel_consumption_l_100km'] <= 30]
```

```python
# Saving new clean csv 
og_data1.to_csv('clean_data3.csv', index=False)
```

```python
og_data1.count()
```

```python
og_data_5y= og_data1.loc[(og_data1['year'] == "2019")|
                       (og_data1['year'] == "2020")|
                       (og_data1['year'] == "2021") | 
                        (og_data1['year'] == "2022") |
                        (og_data1['year'] == "2023")]
```

```python
print(og_data_5y['year'].unique())
```

```python
og_data_5y.to_csv('clean_data_5y.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'year' column to numeric data type (if it's not already)
og_data_5y['year'] = pd.to_numeric(og_data_5y['year'], errors='coerce')

# Group the data by 'year' and 'brand' to calculate the number of registrations for each brand in each year
brand_registrations = og_data_5y.groupby(['year', 'brand']).size().reset_index(name='registrations')

# Sort the data in descending order to find the top 3 brands for each year
top_3_brands_by_year = brand_registrations.sort_values(by=['year', 'registrations'], ascending=[True, False]).groupby('year').head(3)

# Plot a pie chart for each year with the top 3 brands
years = top_3_brands_by_year['year'].unique()
for year in years:
    data_for_year = top_3_brands_by_year[top_3_brands_by_year['year'] == year]
    plt.figure()
    plt.pie(data_for_year['registrations'], labels=data_for_year['brand'], autopct='%1.1f%%', startangle=90)
    plt.title(f"Top 3 Car Brands in {year}")
    plt.axis('equal')
    plt.show()
```

```python
market_share = og_data_5y['brand'].value_counts().head(10)

# Plot a pie chart showing the market share of each brand
plt.figure()
plt.pie(market_share, labels=market_share.index, autopct='%1.1f%%', startangle=90)
plt.title("Market Share of Top 10 Brands - Last 5 Years")
plt.axis('equal')
plt.show()
```
